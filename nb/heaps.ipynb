```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
leftist_heap_insert = pd.read_csv("../data/leftist_heap_insert.csv", header=None)
leftist_heap_insert_random = pd.read_csv("../data/leftist_heap_insert_random.csv", header=None)

binary_heap_insert = pd.read_csv("../data/binary_heap_vec_insert.csv", header=None)
binary_heap_insert_random = pd.read_csv("../data/binary_heap_vec_insert_random.csv", header=None)

binomial_heap_insert = pd.read_csv("../data/binomial_heap_insert.csv", header=None)
binomial_heap_insert_random = pd.read_csv("../data/binomial_heap_insert_random.csv", header=None)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))


leftist_heap_insert.plot.line(x=0, y=1, label="Leftist heap", marker='o', ax=ax)
binary_heap_insert.plot.line(x=0, y=1, label="Binary heap", marker='o', ax=ax)

leftist_heap_insert_random.plot.line(x=0, y=1, label="Leftist heap (random)", marker='o', ax=ax)
binary_heap_insert_random.plot.line(x=0, y=1, label="Binary heap (random)", marker='o', ax=ax)


binomial_heap_insert.plot.line(x=0, y=1, label="Binomial heap", marker='o', ax=ax)
binomial_heap_insert_random.plot.line(x=0, y=1, label="Binomial heap (random)", marker='o', ax=ax)


ax.set_title("Heap insert")
ax.set_ylabel("Time [seconds]")
ax.set_xlabel("N")

ax.grid()
# plt.yscale('log')
#plt.xscale('log')
```
